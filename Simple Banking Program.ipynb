```python
import maskpass
class Bankaccount:
    def __init__(self,account_balance):
        self.account_balance=account_balance
        print("Your bank account balance is:",self.account_balance)

    def deposit(self,amount):
        self.account_balance += amount
        print("Amount Deposited:",amount)
        print("Deposit successful. Your bank account balance is:", self.account_balance)

    def withdraw(self,amount):
        if self.account_balance >= amount:
            self.account_balance -= amount
            print("Amount Withdrawn:", amount)
            print("Withdraw successful. Your bank account balance after withdrawal is:", self.account_balance)
        else:
            print("Insufficient balance, withdraw not possible ")

    def display(self):
        print("Net Available Balance=",self.account_balance)

print("Welcome to ABC Bank ATM")
print("""
Enter 1 for cash deposit
Enter 2 for cash withdrawal
Enter 3 for Checking bank balance
Enter 4 for End
""")

a=int(input("Enter your option:"))
if a==1:
    pin = maskpass.advpass(prompt="Enter your pin:", mask="#", ide=True)
    c=int(pin)
    if c==9999: #should be connected to updated data base of bank here
        obj=Bankaccount(2000) #should be connected to updated data base of bank here
        b=int(input("Enter the amount to deposit which should be a natural number:"))
        obj.deposit(b)
    else:
        print("Incorrect pin number, deposit not possible. Please take back your card and insert again")
        #atm should end transaction procedures, activate card withdrawal option here and restart the process
elif a==2:
    pin = maskpass.advpass(prompt="Enter your pin:", mask="#", ide=True)
    c = int(pin)
    if c == 9999: #should be connected to updated data base of bank here
        obj = Bankaccount(2000) #should be connected to updated data base of bank here
        b = int(input("Enter the amount to withdraw which should be a natural number:"))
        obj.withdraw(b)
    else:
        print("Incorrect pin number, withdraw not possible. Please take back your card and insert again")
        #atm should end transaction procedures, activate card withdrawal option here and restart the process
elif a==3:
    pin = maskpass.advpass(prompt="Enter your pin:", mask="#", ide=True)
    c = int(pin)
    if c == 9999: #should be connected to updated data base of bank here
        obj = Bankaccount(2000)  # should be connected to updated data base of bank here
        obj.display()
    else:
        print("Incorrect pin number, withdraw not possible. Please take back your card and insert again")
        # atm should end transaction procedures, activate card withdrawal option here and restart the process
elif a==4:
    print("Please take back your card and insert again")
    # atm should end transaction procedures, activate card withdrawal option here and restart the process
else:
    print("Invalid entry, please take back your card and insert again")
    # atm should end transaction procedures, activate card withdrawal option here and restart the process
```
